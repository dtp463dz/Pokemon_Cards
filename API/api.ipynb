## Lấy API 

```python
import requests

class Card:
    base_url = "https://api.pokemontcg.io/v2/cards/"

    @classmethod
    def find(cls, card_id):
        # Tạo URL cho card dựa trên ID
        url = cls.base_url + card_id
        # Gửi yêu cầu GET tới API
        response = requests.get(url)
        
        # Kiểm tra mã trạng thái phản hồi
        if response.status_code == 200:
            # Lấy dữ liệu JSON từ phản hồi
            data = response.json()
            return data
        else:
            print(f"Error: {response.status_code}")
            return None

# Tìm thẻ Pokemon có ID 'xy1-1'
card = Card.find('sv7-1')

# Hiển thị thông tin thẻ nếu tìm thấy
if card:
    print(card)
```

```python
# Hiển thị thông tin giá của holofoil từ TCGPlayer
if card:
    holofoil_prices = card['data']['tcgplayer']['prices']['holofoil']
    print("Holofoil Prices:")
    print(f"Low: {holofoil_prices['low']}")
    print(f"Mid: {holofoil_prices['mid']}")
    print(f"High: {holofoil_prices['high']}")
    print(f"Market: {holofoil_prices['market']}")
```

```python
if card:
    cardmarket_prices = card['data']['cardmarket']['prices']
    print("Card Market Prices:")
    print(f"Average Sell Price: {cardmarket_prices['averageSellPrice']}")
    print(f"Low Price: {cardmarket_prices['lowPrice']}")
    print(f"Trend Price: {cardmarket_prices['trendPrice']}")
    print(f"German Pro Low: {cardmarket_prices['germanProLow']}")
    print(f"Suggested Price: {cardmarket_prices['suggestedPrice']}")
    print(f"Average Sell Price (Last 1 Day): {cardmarket_prices['avg1']}")
    print(f"Average Sell Price (Last 7 Days): {cardmarket_prices['avg7']}")
    print(f"Average Sell Price (Last 30 Days): {cardmarket_prices['avg30']}")
```

```python
import requests
import csv

class Card:
    base_url = "https://api.pokemontcg.io/v2/cards?name"

    @classmethod
    def get_all_cards(cls):
        response = requests.get(cls.base_url)
        if response.status_code == 200:
            data = response.json()
            return data
        else:
            print(f"Error: {response.status_code}")
            return None

# Lấy tất cả thông tin thẻ
cards = Card.get_all_cards()

# Tạo file CSV và viết dữ liệu
with open('holofoil_prices_1.csv', mode='w', newline='', encoding='utf-8') as csv_file:
    fieldnames = ['id', 'name', 'low', 'mid', 'high', 'market']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    writer.writeheader()  # Ghi tiêu đề cột

    # Kiểm tra dữ liệu và ghi vào file
    if cards and "data" in cards:
        for card in cards["data"]:
            if ('tcgplayer' in card and 
                'prices' in card['tcgplayer'] and 
                'holofoil' in card['tcgplayer']['prices']):
                
                holofoil_price = card['tcgplayer']['prices']['holofoil']
                
                # Kiểm tra xem các trường giá có tồn tại không
                if all(key in holofoil_price for key in ['low', 'mid', 'high', 'market']):
                    card_data = {
                        'id': card['id'],  # Thêm dấu phẩy ở đây
                        'name': card['name'],
                        'low': holofoil_price['low'],
                        'mid': holofoil_price['mid'],
                        'high': holofoil_price['high'],
                        'market': holofoil_price['market']
                    }
                    writer.writerow(card_data)  # Ghi dữ liệu thẻ vào file

print("Data exported to holofoil_prices_1.csv successfully.")
```

```python
import requests
import csv

class Card:
    base_url = "https://api.pokemontcg.io/v2/cards"
    
    @classmethod
    def get_all_cards(cls):
        response = requests.get(cls.base_url)
        if response.status_code == 200:
            data = response.json()
            return data['data']  # Trả về danh sách thẻ từ API
        else:
            print(f"Error: {response.status_code}")
            return None

# Hàm lưu dữ liệu vào file CSV
def save_to_csv(cards, filename='cards.csv'):
    # Xác định các trường cần lưu
    fields = [
        'id', 'name', 'supertype', 'subtypes', 'hp', 'types', 'evolvesFrom', 'attacks',
        'weaknesses', 'resistances', 'set', 'rarity', 'nationalPokedexNumbers',
        'tcgplayer_holofoil_low', 'tcgplayer_holofoil_mid', 'tcgplayer_holofoil_high',
        'tcgplayer_reverseHolofoil_low', 'tcgplayer_reverseHolofoil_mid', 'tcgplayer_reverseHolofoil_high',
        'cardmarket_averageSellPrice', 'cardmarket_lowPrice', 'cardmarket_trendPrice'
    ]
    
    # Mở file CSV để ghi dữ liệu
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=fields)
        writer.writeheader()  # Ghi tiêu đề cột
        
        for card in cards:
            # Lấy thông tin giá từ tcgplayer
            tcgplayer_prices = card.get('tcgplayer', {}).get('prices', {})
            holofoil = tcgplayer_prices.get('holofoil', {})
            reverse_holofoil = tcgplayer_prices.get('reverseHolofoil', {})
            
            # Lấy thông tin giá từ cardmarket
            cardmarket_prices = card.get('cardmarket', {}).get('prices', {})

            # Viết từng thẻ vào file CSV
            writer.writerow({
                'id': card.get('id'),
                'name': card.get('name'),
                'supertype': card.get('supertype'),
                'subtypes': ', '.join(card.get('subtypes', [])),
                'hp': card.get('hp'),
                'types': ', '.join(card.get('types', [])),
                'evolvesFrom': card.get('evolvesFrom'),
                'attacks': ', '.join([atk['name'] for atk in card.get('attacks', [])]),
                'weaknesses': ', '.join([wk['type'] for wk in card.get('weaknesses', [])]),
                'resistances': ', '.join([res['type'] for res in card.get('resistances', [])]),
                'set': card.get('set', {}).get('name'),
                'rarity': card.get('rarity'),
                'nationalPokedexNumbers': ', '.join(map(str, card.get('nationalPokedexNumbers', []))),
                # Thêm giá tcgplayer holofoil
                'tcgplayer_holofoil_low': holofoil.get('low'),
                'tcgplayer_holofoil_mid': holofoil.get('mid'),
                'tcgplayer_holofoil_high': holofoil.get('high'),
                # Thêm giá tcgplayer reverseHolofoil
                'tcgplayer_reverseHolofoil_low': reverse_holofoil.get('low'),
                'tcgplayer_reverseHolofoil_mid': reverse_holofoil.get('mid'),
                'tcgplayer_reverseHolofoil_high': reverse_holofoil.get('high'),
                # Thêm giá cardmarket
                'cardmarket_averageSellPrice': cardmarket_prices.get('averageSellPrice'),
                'cardmarket_lowPrice': cardmarket_prices.get('lowPrice'),
                'cardmarket_trendPrice': cardmarket_prices.get('trendPrice')
            })

# Lấy dữ liệu từ API
cards = Card.get_all_cards()

# Lưu dữ liệu thành file CSV
if cards:
    save_to_csv(cards)
    print("Data has been saved to 'cards.csv'.")
else:
    print("No data to save.")
```
