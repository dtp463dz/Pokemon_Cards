```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu từ file CSV
df = pd.read_csv('/content/pokemon_cards1.csv')

# Đếm các giá trị trong cột 'rarity'
rarity_counts = df['rarity'].value_counts()

# In ra kết quả
print(rarity_counts)
```

```python
sorted_df = df.sort_values(by='tcgplayer_reverseHolofoil_market', ascending=False)
print(sorted_df[['name', 'tcgplayer_reverseHolofoil_market','rarity']])
```

```python

# Chọn cột cần thiết
df['tcgplayer_holofoil_high'] = pd.to_numeric(df['tcgplayer_holofoil_high'], errors='coerce')

# Nhóm dữ liệu theo 'rarity' và tính giá cao nhất
rarity_price = df.groupby('rarity')['tcgplayer_holofoil_high'].max().reset_index()

# Sắp xếp theo giá cao nhất
rarity_price = rarity_price.sort_values(by='tcgplayer_holofoil_high', ascending=False)

# Vẽ biểu đồ
plt.figure(figsize=(12, 6))
sns.barplot(x='tcgplayer_holofoil_high', y='rarity', data=rarity_price, palette='viridis')
plt.title('Giá cao nhất theo loại thẻ (Rarity)')
plt.xlabel('Giá cao nhất (USD)')
plt.ylabel('Loại thẻ (Rarity)')
plt.xticks(rotation=45)
plt.grid(axis='x')
plt.show()

# In ra bảng giá cao nhất theo rarity
print(rarity_price)
```

#Nhận xét

Loại thẻ Rare Holo có giá trị cao nhất, vì đây là loại thẻ hiếm, được săn lùng rất nhiều bởi những người chơi và người sưu tầm thẻ bài, thêm nữa là giao diện của loại thẻ này rất bắt mắt với những chi tiết lấp lánh làm thu hút mắt người chơi nên được săn đuổi nhiều hơn

```python
# Nhóm dữ liệu theo rarity và tính giá trung bình của tcgplayer_holofoil_high
average_price_by_rarity = df.groupby('rarity')['tcgplayer_holofoil_high'].mean().reset_index()
average_price_by_rarity = average_price_by_rarity.sort_values(by='tcgplayer_holofoil_high', ascending=False)

# Vẽ biểu đồ
plt.figure(figsize=(12, 6))
sns.barplot(x='tcgplayer_holofoil_high', y='rarity', data=average_price_by_rarity, palette='Blues')
plt.title('Giá trung bình cao nhất theo loại thẻ (Rarity)')
plt.xlabel('Giá trung bình cao nhất (USD)')
plt.ylabel('Loại thẻ (Rarity)')
plt.xticks(rotation=45)
plt.grid(axis='x')
plt.show()

print(average_price_by_rarity)
```

#nhận xét


Giá trung bình của loại thẻ Rare Holo không có sự chênh lệch quá nhiều so với các loại thẻ khác, mặc dù ở giá cao nhất thì thẻ Rare Holo có giá trị max khá cao nhưng trong tổng giá trị thì cũng có những giá khá thấp. Chính vì thế sự chênh lệch ở giá trung bình là cao so với loại Rare đứng ở top 2

```python
# Nhóm theo loại thẻ và tính giá trung bình cao nhất
average_price_by_type = df.groupby('types')['tcgplayer_holofoil_high'].mean().reset_index()
average_price_by_type = average_price_by_type.sort_values(by='tcgplayer_holofoil_high', ascending=False)

# Vẽ biểu đồ
plt.figure(figsize=(12, 6))
sns.barplot(x='tcgplayer_holofoil_high', y='types', data=average_price_by_type, palette='Reds')
plt.title('Giá trung bình cao nhất theo loại thẻ')
plt.xlabel('Giá trung bình cao nhất (USD)')
plt.ylabel('Loại thẻ')
plt.xticks(rotation=45)
plt.grid(axis='x')
plt.show()

print(average_price_by_type)
```

Hệ có giá trung bình cao nhất là Psychic,Metal là  220$, về hệ của Pokemon cũng có ảnh hưởng phần nào đến giá trị của thẻ bài, khi có những người chơi muốn sưu tầm theo hệ của pokemon nhưng số lượng thẻ bài không đủ đáp ứng nhu cầu, điều này dẫn đến giá trị thẻ bài được đấy lên cao vì số lượng cung thấp hơn cầu.
